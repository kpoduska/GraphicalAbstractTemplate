```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

from matplotlib.offsetbox import (TextArea, OffsetImage, AnnotationBbox)
from matplotlib.gridspec import GridSpec
```

```python
# Load the data into memory
wnPAS, iPAS = np.loadtxt('calcite_PAS.dpt', delimiter=',', unpack=True)
wnTA, iTA = np.loadtxt('calcite_TA.dpt', delimiter=',', unpack=True)
```

```python
fig = plt.figure(constrained_layout=False)
gs = fig.add_gridspec(nrows=2, ncols=3, left=0, right=1, wspace=0.05, hspace=0.5)

ax1 = fig.add_subplot(gs[:,0], frame_on=False, yticks=[], xticks=[])
img = mpimg.imread('calcite_cell.png')
ax1 = ax1.imshow(img)

ax3 = fig.add_subplot(gs[1,2], yticks=[])
ax2 = fig.add_subplot(gs[0,2], yticks=[], sharex=ax3)


ax3.plot(wnTA, iTA, 'k-', label='Absorbance')
ax2.plot(wnPAS, iPAS, 'k-', label='Photoacoustic')

# Set x and y limits on plot axes (ax3 and ax4)
for ax in (ax3, ax2):    
    ax.set_ylim(0, 1.3*ax.get_ylim()[1])
    ax.set_xlim(2000,600)
    ax.set_xlabel('Wavenumber (cm$^{-1}$)')    
    

bbox_props = dict(boxstyle="rarrow,pad=0.5", fc="w", ec="k", lw=2)


ax.annotate('photoacoustic',
            xy=(150, 210),
            xycoords='figure pixels',
            horizontalalignment='left',
            verticalalignment='top',
            rotation = 20,
            size = 15,
            bbox=bbox_props)

ax.annotate('absorbance',
            xy=(150, 130),
            xycoords='figure pixels',
            horizontalalignment='left',
            verticalalignment='top',
            rotation = -20,
            size = 15,
            bbox=bbox_props)
    
# Draw lines for all band locations
#bandLocs = [(712, 0, 0.65), (845, 0, 0.8), (875, 0, 0.75), (1425, 0, 0.9), (1795, 0, 0.45), (2520, 0, 0.36), (2870, 0, 0.25)]

#for bandLoc in bandLocs:
#    for ax in (ax3, ax4):
#        ax.axvline(x=bandLoc[0], ymin=bandLoc[1], ymax=bandLoc[2], linestyle=':', linewidth=2, c='k')

plt.show()
```
